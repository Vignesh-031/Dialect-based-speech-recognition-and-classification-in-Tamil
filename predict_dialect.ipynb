```python
import os
import whisper
import pandas as pd
from transformers import pipeline

# =========================
# LOAD MODELS
# =========================
#speech_model = whisper.load_model("base")

classifier = pipeline(
    "text-classification",
    model="./final_dialect_model",
    tokenizer="l3cube-pune/tamil-bert"
)

label_reverse = {0:"C", 1:"N", 2:"S", 3:"W"}

# =========================
# AUDIO FOLDER PATH
# =========================
folder_path = r"C:\Users\Student\Desktop\Dialet\Dataset\Central_Dialect\SP10_THA_audio"

results = []

# =========================
# LOOP THROUGH FILES
# =========================
for file in os.listdir(folder_path):
    if file.lower().endswith((".wav", ".mp3", ".m4a")):

        audio_path = os.path.join(folder_path, file)

        print(f"\nProcessing: {file}")

        # ---- Speech → Text ----
        result = speech_model.transcribe(audio_path, language="ta")
        text = result["text"]

        print("Text:", text)

        # ---- Text → Dialect ----
        pred = classifier(text)[0]
        dialect = label_reverse[int(pred["label"].split("_")[-1])]

        print("Dialect:", dialect)

        results.append({
            "file": file,
            "text": text,
            "predicted_dialect": dialect
        })

# =========================
# SAVE RESULTS
# =========================
df = pd.DataFrame(results)
df.to_csv("audio_predictions.csv", index=False)

print("\n✅ Done! Results saved to audio_predictions.csv")
```
